```python
import pandas as pd

try:
    titanic_df = pd.read_csv('Titanic-Dataset.csv')
    display(titanic_df.head())
except FileNotFoundError:
    print("Error: 'Titanic-Dataset.csv' not found.")
except pd.errors.ParserError:
    print("Error: Unable to parse the CSV file.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

## Data exploration

### Subtask:

```python
# Determine the shape of the DataFrame.
print("Shape of the DataFrame:", titanic_df.shape)

# Determine the data type of each column.
print("\nData types of each column:\n", titanic_df.dtypes)

# Calculate the number of missing values in each column.
print("\nNumber of missing values in each column:\n", titanic_df.isnull().sum())
```

## Data analysis

### Subtask:
Calculate summary statistics and correlations for numerical features.


**Reasoning**:
Calculate summary statistics for numerical features and display the correlation matrix with a heatmap.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select numerical features.
numerical_features = ['PassengerId', 'Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
numerical_df = titanic_df[numerical_features]

# Calculate and display summary statistics.
print("Summary Statistics:\n", numerical_df.describe())

# Calculate and display the correlation matrix.
correlation_matrix = numerical_df.corr()
print("\nCorrelation Matrix:\n", correlation_matrix)

# Visualize the correlation matrix using a heatmap.
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

## Data visualization

### Subtask:
Visualize numerical features using histograms and box plots, and visualize relationships between features using a pairplot.


**Reasoning**:
Visualize numerical features using histograms, box plots, and a pairplot to understand their distributions and relationships.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Numerical features to visualize
numerical_features = ['Age', 'Fare', 'SibSp', 'Parch']

# Histograms
plt.figure(figsize=(12, 8))
for i, feature in enumerate(numerical_features):
    plt.subplot(2, 2, i + 1)
    sns.histplot(titanic_df[feature].dropna(), kde=True)  # Drop NaN values for histograms
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# Box plots
plt.figure(figsize=(12, 8))
for i, feature in enumerate(numerical_features):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(y=titanic_df[feature])
    plt.title(f'Box Plot of {feature}')
    plt.ylabel(feature)
plt.tight_layout()
plt.show()

# Pairplot
plt.figure(figsize=(10, 10))
sns.pairplot(titanic_df[numerical_features].dropna()) # Drop NaN values for pairplot
plt.suptitle('Pairplot of Numerical Features', y=1.02) # Added title for the pairplot
plt.show()
```

## Data visualization

### Subtask:
Visualize categorical features (Survival rate by Passenger class, Sex, Embarked) using bar charts or count plots.


**Reasoning**:
Visualize the survival rate by passenger class, sex, and embarked location using bar plots. Handle missing values in 'Embarked' before plotting.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Fill missing values in 'Embarked' with the most frequent value
titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0], inplace=True)

# Survival rate by passenger class
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', hue='Survived', data=titanic_df, palette='viridis')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.show()

# Survival rate by sex
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', hue='Survived', data=titanic_df, palette='magma')
plt.title('Survival Rate by Sex')
plt.xlabel('Sex')
plt.ylabel('Number of Passengers')
plt.show()

# Survival rate by embarked location
plt.figure(figsize=(8, 6))
sns.countplot(x='Embarked', hue='Survived', data=titanic_df, palette='plasma')
plt.title('Survival Rate by Embarked Location')
plt.xlabel('Embarked Location')
plt.ylabel('Number of Passengers')
plt.show()
```
